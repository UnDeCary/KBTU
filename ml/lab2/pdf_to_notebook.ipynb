```python
import pdfplumber
import nbformat
from nbformat.v4 import new_notebook, new_markdown_cell, new_code_cell
import re

PDF_FILE = "tasks.pdf"
OUTPUT_FILE = "assignment.ipynb"

# Регулярка для вопросов
question_pattern = re.compile(r".*\d+\.\d+\.\d+\s+.*")

cells = []

# 1️⃣ Читаем PDF и собираем текст
with pdfplumber.open(PDF_FILE) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if not text:
            continue

        # 2️⃣ Обрабатываем текст построчно (как ass.txt)
        for line in text.splitlines():
            line = line.strip()
            

            if question_pattern.match(line):
                # Markdown с вопросом
                cells.append(
                    new_markdown_cell(f"#### {line}")
                )
                # Пустая кодовая ячейка под решение
                cells.append(
                    new_code_cell("")
                )

# 3️⃣ Создаём ноутбук
nb = new_notebook(cells=cells)

# 4️⃣ Сохраняем .ipynb
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    nbformat.write(nb, f)

print(f"Notebook saved as {OUTPUT_FILE}")
```
