# Q1: What is the Probability that the Customer has a Credit Card?

```python
total_customers = df["CreditCard"].count()
has_credit_card = df[df["CreditCard"] == "yes"]["CreditCard"].count()
print(round((has_credit_card / total_customers) * 100, 3), "%")
```

# Q2: What is the Probability that the Customer age is greated than 50?

```python
print(round((df["age"] > 50).mean() * 100, 3), "%")
```

# Q3. What is the probability that the Customer is illiterate?

```python
print( round( ( df["education"] == "illiterate").mean() * 100, 10 ), "%")
```

# Q4: What is the probability that the Customer has a Telephone?

```python
print(round((df["contact"] == "telephone").mean() * 100, 3), "%")
```

# Q5: What is the Probability that the Customer is Singe and has a Credit Card?

```python
print(round(((df.marital == "single") & (df.CreditCard == "yes")).mean() * 100, 3), "%")
```

# Q6: What is the probability that the Customer is Male and has at most 6 Year of Education?

```python
has_at_most_6_y = (df.education.isin(["illiterate", "unknown", "basic.4y", "basic.6y"]))
is_male = (df.gender == "male")
is_male_and_at_most_6_y = (has_at_most_6_y & is_male).mean()
print(round(is_male_and_at_most_6_y * 100, 3), "%")
```

# Q7: What is the probability that the Customer is Retired and is divorced?

```python
is_retired = df.job == "retired"
is_divorced = df.marital == "divorced"
print(round((is_retired & is_divorced).mean() * 100, 3), "%")
```

# Q8: Given the Customer has the credit card, find the probability that he/she is married?

```python
has_card_and_married = (df[df["CreditCard"] == "yes"]["marital"] == "married").mean()
print(round(has_card_and_married.mean() * 100,3), "%")
```

# Q9: If a customer has a management job, find the probability that he has a housing?

```python
managment_job_and_housing = (df[df["job"] == "management"]["housing"] == "yes").mean()
print(round(managment_job_and_housing.mean() * 100,3), "%")
```

```python
p_june_more_500 = df['JUNE_MORE_THAN_500'].mean()
p_flood = df['IS_FLOODS'].mean()
p_flood_if_june_more_500 = df[df['JUNE_MORE_THAN_500'] == 1]['IS_FLOODS'].mean()

p_june_more_500_if_flood = (p_flood_if_june_more_500 * p_june_more_500) / p_flood
print(p_june_more_500_if_flood)
```

```python
p_june_more_500 = df['JUNE_MORE_THAN_500'].mean()
p_june_less_500 = 1 - p_june_more_500
p_flood_june_more_500 = df[df['JUNE_MORE_THAN_500'] == 1]['IS_FLOODS'].mean() 
p_flood_june_less_500 = df[df['JUNE_MORE_THAN_500'] == 0]['IS_FLOODS'].mean()

p_june_more_500_if_flood = (p_flood_june_more_500 * p_june_more_500) / ((p_flood_june_more_500 * p_june_more_500) + (p_flood_june_less_500 * p_june_less_500))
print(p_june_more_500_if_flood)
```
