# Q2 Given vector a = [5, -3] and scalar $\lambda$ = -2

```python
a = np.array([5, -3])
lambda_ = -2

# 1. Compute $\lambda$ a

result = lambda_ * a
print(f"lambda * a = {result}")

# 2. Compare their magnitudes
print(f"||a|| = {np.linalg.norm(a)}")
print(f"||result|| = {np.linalg.norm(result)}")
```


#### 3. Explain the effect on magnitude and direction.
1. magnitude multiplies by |$\lambda$|
2. direction change if $\lambda$ < 0

# Q5. Given vector a = [1,2,3,4,5,6], compute its Euclidean norm manually and using NumPy.


```python
a = np.array([1, 2, 3, 4, 5, 6])

c_sum = 0

for i in a:
    c_sum += i**2
print(c_sum**0.5)

manual = np.sqrt(np.sum(a**2))
print(manual)

norm = np.linalg.norm(a)
print(norm)
```

# Q6. Generate a random vector of length 10 and normalize it. Print the norm before and after normalization

```python
v = np.random.rand(10)

before_norm = np.linalg.norm(v)

v_norm = v / before_norm

after_norm = np.linalg.norm(v_norm)
print(before_norm, after_norm, sep='\n')
```

# Q7. Compute the dot product of vectors

```python
a = [1, 2, 3, 4, 5, 6]
b = [0, -4, 3, -3, 6, 5]

# Manual loop
man_res = 0
for i in range(len(a)):
    man_res += a[i] * b[i]
print(man_res)

np.dot(a, b)
```

# Q8. Verify the distributive property:

```python
a = np.random.rand(3)
b = np.random.rand(3)
c = np.random.rand(3)

left = np.dot(a, b + c)
right = np.dot(a, b) + np.dot(a, c)
print(left, right, sep=' = ')
```

# Q9: Compute the Euclidean distance matrix between every point in group A and every point in group B.

```python
# Group A
A = np.array([
 [ 2, 1],
 [ 5, 4],
 [ 1, 6],
 [ 7, 2]
]) 
# Group B
B = np.array([
 [0, 0],
 [6, 3],
 [3, 5]
])

A_n = A[:, None, :]
B_n = B[None, :, :]

dist = np.sqrt(((A_n - B_n) ** 2).sum(axis=2))
print(A_n - B_n)
dist
```
